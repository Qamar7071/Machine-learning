# Libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import root_mean_squared_error, mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
dataset='Student_performance_data _.csv'
df=pd.read_csv(dataset)
print(df)
df.columns
```

```python
X=df[['StudyTimeWeekly']]
X2=df[['StudyTimeWeekly','Absences','ParentalSupport']]
y=df['GPA']
y2 = df['GradeClass']

model=LinearRegression()
model.fit(X,y)
model2 = LogisticRegression()
model2.fit(X2, y2)
predicted_score=model.predict(X)
print(predicted_score)
predicted_Grade=model2.predict(X2)
print(f"Predicted Grade: {predicted_Grade}")
r2=r2_score(y,predicted_score)
r3=r2_score(y,predicted_Grade)
rmse = np.sqrt(mean_squared_error(y, predicted_score))
mae=mean_absolute_error(y,predicted_score)
mse=mean_squared_error(y,predicted_score)
print("mean_absolute_error",round(mae,2))
print("mean_squared_error",round(rmse,2))
print("r2_score",round(r2,4))
print("square_root",round(rmse,2))

print("r2_score_Grade",round(r3,4))

```

# Scatter and Regression line

```python
# Plot
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual Score')
plt.plot(X, predicted_score, color='green', label='Predicted GPA (Regression line)')
plt.title('Model Prediction vs Actual Score')
plt.xlabel('Study hours per week')
plt.ylabel('GPA')
plt.legend()
plt.grid(True)
plt.show()
hours=9
predicted_score=model.predict([[hours]])
print(f"predicted final score of {predicted_score}" )
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual Score')
plt.plot(X, predicted_Grade, color='blue', label='Predicted Grade (Regression line)')
plt.title('Model Prediction vs Actual Score')
plt.xlabel('Study hours per week')
plt.ylabel('Grade')
plt.legend()
plt.grid(True)
plt.show()
hours=10
predicted_Grade=model2.predict([[hours]])
print(f"predicted final Grade of {predicted_Grade}" )
```
