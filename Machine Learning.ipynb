```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
data="messy_employee_data.csv"
pf=pd.read_csv(data)
print(pf.head())
print(pf.shape)
print(pf.isnull().sum())
print(pf.dropna())
pf['Age'] = pf['Age'].fillna(pf['Age'].mean())
pf['Salary']=pf['Salary'].fillna(pf['Salary'].mean())
print(pf.isnull().sum())
pf['Salary']=np.where(pf['Salary']>0,pf['Salary'],pf['Salary'].mean())
print(pf['Salary'])

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('Box Plot of salary vs Age')
plt.boxplot(pf['Salary'])
plt.ylabel('Salary')
```

```python

pf=pd.get_dummies(pf[['Department','Performance_Rating']]).astype(int)
std_scaler=StandardScaler()
std_scaled=std_scaler.fit_transform(pf)
print('std_scaled')
print(std_scaled)
```
