```python
class Person:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)


class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade
    def display(self):
        print("Name:", self.name)
        print("Grade:", self.grade)


s1 = Student("Ali", "A")
s2=Student("Qamar", "B")
s3=Student("Arfa", "C")
s1.display()

s1.show()
s1.display()
s2.show()
s3.display()
```

```python
class cal:
    def add(self, a, b, c=0):
        return a + b + c

st = cal()
print(st.add(1,2))
print(st.add(1,2,3))
```
